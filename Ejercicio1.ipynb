```python
import numpy as np
import matplotlib.pyplot as plt

n_pedidos = 20
p_exito = 0.7
dias = 7

pedidos_satisfechos = np.random.binomial(n_pedidos, p_exito, dias)

print("Pedidos satisfechos por día:", pedidos_satisfechos)

plt.bar(range(1, dias + 1), pedidos_satisfechos, color='royalblue', edgecolor='black')
plt.xticks(range(1, dias + 1))
plt.xlabel("Día")
plt.ylabel("Pedidos Satisfechos")
plt.title("Simulación de Pedidos Satisfechos por Día")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#Conclusión
####Las variables aleatorias ayudan a representar la incertidumbre en los tiempos de servicio, permitiendo un análisis más realista del sistema. Su uso mejora la gestión de recursos, detecta cuellos de botella y optimiza la asignación para reducir esperas y mejorar la calidad del servicio. También contribuyen a disminuir costos al prever problemas y ajustar la capacidad operativa.

La simulación con variables aleatorias permite probar distintos escenarios sin afectar el sistema real, facilitando decisiones estratégicas y mejorando la planificación. Es clave en sectores como logística, salud y manufactura, donde la eficiencia y la predicción son esenciales para el éxito.
