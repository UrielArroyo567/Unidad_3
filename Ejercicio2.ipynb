```python
import numpy as np
import matplotlib.pyplot as plt

tasa_llegadas = 1 / 15
horas_operacion = 12
minutos_operacion = horas_operacion * 60
num_llegadas =  np.random.poisson(tasa_llegadas * minutos_operacion)

tiempos_entre_llegadas = np.random.exponential(scale=1/tasa_llegadas, size=num_llegadas)

print("Tiempos entre llegadas (minutos):", tiempos_entre_llegadas)

plt.hist(tiempos_entre_llegadas, bins=20, color='royalblue', edgecolor='black', density=True)
plt.xlabel("Tiempo entre llegadas (minutos)")
plt.ylabel("Densidad")
plt.title("Distribuci√≥n de tiempos entre llegadas")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#Conclusi√≥n
####Simular los tiempos entre llegadas de veh√≠culos el√©ctricos con una distribuci√≥n exponencial permite modelar de forma realista la variabilidad en la demanda de carga. Esto ayuda a prever la cantidad de puntos de carga necesarios para reducir esperas y mejorar el flujo de atenci√≥n.

Este enfoque optimiza la planificaci√≥n de recursos, ajustando la capacidad operativa seg√∫n la demanda. Adem√°s, al identificar patrones en las llegadas, se pueden tomar decisiones estrat√©gicas para aumentar la eficiencia del servicio y mejorar la experiencia del usuario.
