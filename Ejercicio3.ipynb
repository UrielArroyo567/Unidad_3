```python
import numpy as np
import matplotlib.pyplot as plt

media_espera = 10
desviacion_espera = 2
num_clientes = 50

mu = np.log(media_espera) - 0.5 * np.log(1 + (desviacion_espera / media_espera) ** 2)
sigma = np.sqrt(np.log(1 + (desviacion_espera / media_espera) ** 2))

tiempos_espera = np.random.lognormal(mean=mu, sigma=sigma, size=num_clientes)

print("Tiempos de espera (minutos):", tiempos_espera)

plt.hist(tiempos_espera, bins=15, color='royalblue', edgecolor='black', density=True)
plt.xlabel("Tiempo de espera (minutos)")
plt.ylabel("Densidad")
plt.title("Distribución de tiempos de espera en servicio técnico")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#Conclusión
####Simular los tiempos de espera con una distribución lognormal permite representar con precisión la variabilidad en la atención del servicio técnico. Esto ayuda a prever momentos de alta demanda y mejorar la gestión de filas.  

Con esta información, la tienda puede optimizar recursos, reducir la congestión en horas pico y ofrecer un servicio más eficiente, minimizando esperas prolongadas y mejorando la experiencia del cliente.
